```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
```

```python
teacher_details = {
    'ID_repetitor': [],
    'price': [],
    'experience': [],
    'age': [],
    'reviews': [],
    'education': [],
    'video': [],
    'photo': [],
    'diplomas': [],
    'students': [],
    'years' : [],
    'last_activity': [],
    'chapters': [],
    'additions': [],
    'home': [],
    'out': [] 
    }

start_time = time.time()

for page in range(1, 596):
    url = f'https://repetit.ru/repetitors/matematika/zhenshchini/?lp=1,2&page={page}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    teacher_links = soup.find_all('a', {'class': 'teacher-name'})

    for link in teacher_links:
        teacher_url = "https://repetit.ru" + link['href']
        teacher_response = requests.get(teacher_url)
        teacher_soup = BeautifulSoup(teacher_response.content, 'html.parser')

        ID_repetitor_check = teacher_soup.select_one('span.id') 
        ID_repetitor = ID_repetitor_check.text.strip() if ID_repetitor_check is not None else ''
        price_check = teacher_soup.find('p', {'class': 'price'}).find('strong')
        price = price_check.text.strip() if price_check is not None else ''
        reviews_check = teacher_soup.select_one('.reviews h2 span')
        reviews = reviews_check.text.strip() if reviews_check is not None else ''
        education_check = teacher_soup.find('div', {'class': 'education'})
        education = education_check.text.strip() if education_check is not None else ''
        experience_check = teacher_soup.find('div', {'class': 'experience'}).find('span', {'class': 'hide-in-edit-mode'})
        experience = experience_check.text.strip() if experience_check is not None else ''
        age_check = teacher_soup.select_one('.age div.col-8')
        age = age_check.text.strip() if age_check is not None else ''
        video_check = teacher_soup.select_one('.video div.section-header')
        video = video_check.text.strip() if video_check is not None else 0
        photo = teacher_soup.find('div', {'class': 'teacher-photo'}).find('img')['src']
        diplomas_check = teacher_soup.select_one('.diplomas h2 span')
        diplomas = diplomas_check.text.strip() if diplomas_check is not None else 0
        students_check = teacher_soup.select_one('div.col-3:nth-of-type(4) h5')
        students = students_check.text.strip() if students_check is not None else ''
        years_check = teacher_soup.select_one('.statistic div:nth-of-type(4) p')
        years = years_check.text.strip() if years_check is not None else ''
        last_activity_check = teacher_soup.select_one('div.col-3:nth-of-type(3) h5')
        last_activity = last_activity_check.text.strip() if last_activity_check is not None else ''
                
        chapters = ''
        additions = ''
        subject_divs = teacher_soup.select('div.subject')
        for subject_div in subject_divs:
            subject_h4 = subject_div.select_one('h4')
            if subject_h4 is not None and subject_h4.text.strip() == 'Математика':
                if subject_div.has_attr('class') and 'expand-it' in subject_div['class']:
                    chapters = subject_div.select_one('.expand-it div:nth-of-type(1) div.col-8').text.strip()
                    additions = subject_div.select_one('.expand-it div.row:nth-of-type(2) div.col-8').text.strip()
                else:
                    chapters_divs = subject_div.select('div.col-8')
                    chapters = chapters_divs[0].text.strip()
                    additions = chapters_divs[1].text.strip()
                break
                
        home_check = teacher_soup.select_one('.home-addr-group')
        home = home_check.text.strip() if home_check is not None else ''
        out_check = teacher_soup.select_one('.away-addr-group')
        out = out_check.text.strip() if out_check is not None else ''
        
        
        teacher_details['ID_repetitor'].append(ID_repetitor)
        teacher_details['age'].append(age)
        teacher_details['experience'].append(experience)
        teacher_details['price'].append(price)
        teacher_details['reviews'].append(reviews)
        teacher_details['education'].append(education)
        teacher_details['video'].append(video)
        teacher_details['photo'].append(photo)
        teacher_details['diplomas'].append(diplomas)
        teacher_details['students'].append(students)
        teacher_details['years'].append(years)
        teacher_details['last_activity'].append(last_activity)
        teacher_details['chapters'].append(chapters)
        teacher_details['additions'].append(additions)
        teacher_details['home'].append(home)
        teacher_details['out'].append(out)

end_time = time.time()

print(f"Execution time:\n{(end_time - start_time)/60} minutes = {(end_time - start_time)/3600} hours")
```

```python
df_women = pd.DataFrame(teacher_details)
df_women
```
